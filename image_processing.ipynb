#Data augmentation

```
aug=Image.open('sky.png')
aug.show()
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
gen=ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
# fillmode used when the image is resized or whenever new spaces are generated
```

```
plt.imshow(aug)
```

```
from tensorflow.keras.preprocessing.image import array_to_img, img_to_array, load_img
#converting the image to a numpy equivalent array so that reshaping can be applied
aug=img_to_array(aug)
aug=aug.reshape((1,)+aug.shape)
i = 0
for batch in gen.flow(aug, batch_size=1):
  #flow generates randomly transformed data
  plt.figure(i)
  imgplot = plt.imshow(array_to_img(batch[0]))
  i += 1
  if i % 5 == 0:
    break
```
